```python
import os
import youtube_dl
from tqdm import tqdm, trange
import time
```

```python
import pandas as pd
```

```python
a=pd.read_csv('1_all.csv')
a
```

```python
a=a[a['title']!='삭제된 동영상을 시청함']
a=a[a['title']!='Visited YouTube Music']
```

```python
download_list=a['titleUrl']
download_list=list(download_list)
download_list=set(download_list)
download_list=list(download_list)
download_list
```

```python
real=[]
```

```python
for i in download_list:
    try:
        if 'https://www.youtube.com/watch?v=' in i:
            real.append(i)
    except:
        pass
```

```python
real
```

```python
example=real[10:40]
example
```

```python
for i in tqdm(example):
    output_dir = os.path.join('./', '%(title)s.%(ext)s')
    ydl_opt = {'outtmpl': output_dir}# 실행되는 폴더 안에 '영상제목.확장자' 형식으로 다운로드
    try:
        with youtube_dl.YoutubeDL(ydl_opt) as ydl:
            ydl.download([i])
        print('다운로드 완료했습니다.')
    except:
            pass
```
