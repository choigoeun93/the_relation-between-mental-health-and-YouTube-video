```python
file_path = './regnet_final_087.h5'
batch_size = 200
nb_samples = 2000
SIZE = (224, 224)

model = load_model(file_path)
test_datagen = ImageDataGenerator()
```

```python
test_data_dir = 'C:\\Users\\choig\\Downloads\\18_18'
```

```python
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=SIZE,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False)

print(test_generator)
```
